<h2> Wprowadzenie</h2>

Statystyka i modelowanie statystyczne
R oferuje szeroki zakres metod: od klasycznej regresji liniowej i testów hipotez, przez modele mieszane i analizy wielowymiarowe, po zaawansowane modele bayesowskie. Jego składnia i struktura danych (wektory, ramki danych, listy) są zaprojektowane z myślą o analizie statystycznej.

Data science i eksploracja danych
Dzięki pakietom takim jak dplyr, data.table czy tidyr, **R** umożliwia efektywne przetwarzanie i transformację danych. Jest szeroko wykorzystywany do eksploracyjnej analizy danych, w tym wizualizacji.

Wizualizacja danych
**R** jest szczególnie ceniony za wysoką jakość i elastyczność wizualizacji. 

Bioinformatyka i analiza danych medycznych
Dzięki projektowi <a href="https://www.bioconductor.org">Bioconductor R</a> pozostaje jednym z dominujących narzędzi w analizie danych genomowych, transkryptomicznych i proteomicznych. W środowisku akademickim i badawczym jest to wciąż standard.

Ekonometria i nauki społeczne
**R** jest szeroko stosowany w ekonomii, socjologii i politologii do modelowania danych panelowych, analiz regresyjnych i symulacji. W wielu czasopismach naukowych publikacje opierają się na analizach przeprowadzonych w **R**.

Raportowanie i reproducibility

Mimo rosnącej popularności języków takich jak Python w obszarze ogólnego data science i uczenia maszynowego, **R** nadal utrzymuje silną pozycję tam, gdzie kluczowe są zaawansowane metody statystyczne, wysoka jakość wizualizacji oraz rygor naukowy analizy danych.

Wszystkie one wykonują to samo działanie na najistotniejsze różnica leży w wartościach domyślnych argumentów tych funkcji. Wczytując dane z pliku najważniejsze parametry, które należy wziać pod uwagę to:

- `header` - określa czy plik zawiera nagłówek (pierwszy wiersz z nazwami kolumn). Dopuszczalne wartości to `TRUE` oraz `FALSE`.
- `sep` - znak, który rozdziela kolumny w pliku (najczęściej `','`, `';'` lub `'\t'`).
- `dec` - znak używany w zapisie liczb zmiennoprzecinkowych (najczęściej `'.'` lub `','`).
- `encoding` - pozwala określić kodowanie pliku (najczęściej  `Latin-1` lub `UTF-8`).

Domyśle wartości dla powyższych funkcji to:
- `read.table(file,  header = FALSE, sep = "",   dec = ".", ...)`
- `read.csv(file,    header = TRUE,  sep = ",",  dec = ".", ...)`
- `read.csv2(file,   header = TRUE,  sep = ";",  dec = ",", ...)`
- `read.delim(file,  header = TRUE,  sep = "\t", dec = ".", ...)`
- `read.delim2(file, header = TRUE,  sep = "\t", dec = ",", ...)`

Funkcje te pozwalają określić jeszcze szereg innych informacji o pliku. Proszę przejrzeć dostępne parametry na przykładzie dokumentacji funkcji <a href="https://www.rdocumentation.org/packages/utils/versions/3.6.2/topics/read.table">read.table</a>.

| Obiekt w R                         | typeof()       | class()             |

| `list(1, "a", TRUE)`                | "list"        | "list"             |

Zmienne typu `double`, `integer`, `logical`, `character`, `complex`, `list` i `NULL` nie wymagają chyba komentarza. <h3>factor - zmienne kategoryczne</h3>

Typ `factor` przeznaczony jest dla danych kategorycznych. Opcjonalnie pozwala wprowadzić porządek pomiędzy wartościami nominalnymi. Jego działanie można intuicyjne interpretować jak działanie słownika w Pythonie, który tłumaczy wartości nominalne na odpowiedni indeks. Działanie na intach pozwala przyśpieszyć obliczenia.  

```r
#Przykładowy zbiór danych
ocena <- factor(c("niska", "średnia", "wysoka", "średnia", "niska"))
ocena
```

Dopuszczalne wartości określone są w polu **levels**.  Można  określić również poziomy, które nie wystąpiły we wczytanych danych.

```r
levels(ocena)
```

```r
ocena <- factor(c("niska", "średnia", "wysoka", "średnia", "niska"), levels=c("niska", "średnia", "wysoka","bardzo wysoka"))
ocena
```

```r
levels(ocena)
```

Podsumowanie informacji zawartych w tej zmiennej możemy odczytać funkcją `str`.

```r
str(ocena)
```

Informację o częstości każdego levelu można odczytać funkcją `table`.

```r
table(ocena)
```

O ile zawsze wartością przy porządkowe są "techniczne" inty, to możemy wymusić by system uwzględniał relację między nimi, czyli wiedział które wartości mają być interpretowane jako większe od innych.

```r
ocena_ord <- factor(
  c("niska", "średnia", "wysoka"),
  levels = c("niska", "średnia", "wysoka"),
  ordered = TRUE
)

ocena_ord
```

```r
#Dla danych uporządkowanych - wartość boolowa
ocena_ord[1] < ocena_ord[2]
```

```r
#Dla danych nieuporządkowanych - ostrzeżenie i NA
ocena[1] < ocena[2]
```

W **R** istnieją trzy podstawowe pętle: `for`, `while` i `repeat`, a ich budowa nie różni się zasadniczo od tych znanych z innych języków.

```r
for(i in sekwencja) {
    print(i)}
```

```r
i=1
while(sekwencja[i] < 9) {
    print(sekwencja[i])
    i=i+1}
```

```r
i=1
repeat {
  print(sekwencja[i])
  i = i+1
  if(sekwencja[i]>5) {
    break
  }
}
```

Przy okazji zauważyliśmy, że działa komenda `break`. Podobnie działa również `continue`.

<h2>Instrukcje warunkowe</h2>

**R** dostarcza dwa dwie podstawowe instrukcje warunkowe `if` oraz `switch`.

```r
x = 0
if (x < 0) {
  print("liczba ujemna")
} else if (x > 0) {
  print("liczba dodatnia")
} else
  print("zero")
```

Warto może zwrócić uwagę na sprawdzanie typów zmiennych. Do tego celu służą komendy `is.integer`, `id.double`, itp.

```r
x = 30L
if (is.integer(x))
{
  print("X to Int")
}
```

Podobie do rzutowania zmiennych istnieją funkcje typu `as.integer`, `as.double`, itp.

```r
x = 30
if (is.integer(x))
{
  print("X to Int")
} else
{
  print("X to nie jest Int")    
}
x = as.integer(x)
if (is.integer(x))
{
  print("Już jest intem")
} 
```

Może wyszukiwać wartości w listach korzystając z `%in%`.

```r
x = c("jeden","dwa","trzy")

if("dwa" %in% x) {
  print("Znaleziono")
} else {
  print("Brak")
}
```

Instrukcja `switch` służy do wyboru jednej z kilku opcji. Instrukcja nie przechodzi do kolejnych poniżej.

`switch(EXPR,`

`    name1 = value1,`

`    name2 = value2,`

`    name3 = value3)`

```r
switch(
  1,
  "first",
  "second",
  "third",
  "fourth"
)
```

```r
operacja <- "pomnoz"

wynik <- switch(operacja,
                dodaj = 2 + 2,
                odejmij = 2 - 2,
                pomnoz = 2 * 2,
                "Nieznana operacja")
wynik
```

```r
#wartość domyślna
operacja <- "poteguj"

wynik <- switch(operacja,
                dodaj = 2 + 2,
                odejmij = 2 - 2,
                pomnoz = 2 * 2,
                "Nieznana operacja")
wynik
```

Korzystanie ze zmiennych

```r
x <- 16
fun <- "sqrt"

wynik <- switch(fun,
                sqrt = sqrt,
                log = log,
                exp = exp)(x)
wynik
```

<h2>Funkcje</h2>

Funkcję przypisujemy do zmiennej. Tworzymy ją wyrażeniem `function`.

```r
# Tworzenie funkcji
new.function <- function(a) {
  for(i in 1:a) {
    b <- i^2
    print(b)
  }
}	
```

```r
# wywołanie funkcji
new.function(6)
```

Możemy w różny sposób korzystać z parametrów.

```r
# funkcja bez parametru
new.function <- function() {
  for(i in 1:5) {
    print(i^2)
  }
}	
```

```r
# wywołanie
new.function()
```

```r
# funkcja z parametrami
new.function <- function(a,b,c) {
  result <- a * b + c
  print(result)
}
```

```r
# parametry podane pozycyjnie
new.function(5,3,11)
```

```r
# parametry podane z nazwami
new.function(a = 11, b = 5, c = 3)
```

```r
# utworzenie funkcji z parametrami ze zdefiniowanymi wartościami domyślnymi
new.function <- function(a = 3, b = 6) {
  result <- a * b
  print(result)
}
```

```r
new.function()
```

```r
new.function(9,5)
```

Uwaga! 

Parametr jest sprawdzany dopiero wtedy, kiedy ma być użyty.

```r
new.function <- function(a, b) {
  print(a^2)
  print(a)
  print(b)
}
```

```r
# wywołanie funkcji bez jednego parametru.
new.function(6)
```
