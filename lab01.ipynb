```r
data <- rnorm(mean = 0, sd = 1, n = 100)
data[c(1:5)]
```

```r
# rnorm ()
# dnorm (zwraca wartosc funkcji gestosci),
# pnorm (zwraca wartosc dystrybuanty),
# qnorm (zwraca kwantyl)
```

```r
# Test Shapira-Wilka
shapiro.test(data)

# Hipoteza zerowa: dane pochodza z rozkladu normalnego
# Hipoteze zerowa mozna odrzucic, ale nie mozna zakceptowac (mowi sie ze nie ma podstaw do odrzucenia)
```

```r
chi_data <- rchisq(n = 100, df = 5)
shapiro.test(chi_data)
```

```r
# Zrob to na wektorach (te operacje)

u <- runif(n = 4, min = 0, max = 1)
u1 <- u[1]
u2 <- u[2]
u3 <- u[3]
u4 <- u[4]

z1 <- sqrt(-2 * log(u1)) * cos(2 * pi * u2)
z2 <- sqrt(-2 * log(u1)) * sin(2 * pi * u2)
z3 <- sqrt(-2 * log(u3)) * cos(2 * pi * u4)
z4 <- sqrt(-2 * log(u3)) * sin(2 * pi * u4)

shapiro.test(c(z1, z2, z3, z4))
```
